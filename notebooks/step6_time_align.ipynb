```python
%matplotlib widget
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams
from ipywidgets import IntSlider, interact

rcParams['font.family'] = 'SimHei'
rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv("../data/raw/may_filtered.csv")
datetimestr = df['date'] + ' ' + df['time']
df['datetime'] = pd.to_datetime(datetimestr, format='%d-%m-%Y %H:%M:%S').astype("datetime64[s]")
df['duration(s)'] = pd.to_timedelta(df.duration).dt.total_seconds()
df.drop(['date', 'time', 'duration'], inplace=True, axis=1)
df['motion'] = np.linalg.norm(df[['accX', 'accY', 'accZ']].values, axis=1)
df = df[["datetime", "duration(s)", "red", "ied", "motion"]]
```

```python
time = df.datetime.astype("int64")
di = np.diff(time)
di = np.insert(di, 0, 0)
time_di = pd.Series(di)
time_di_freq = time_di.value_counts()
```

```python
dur = df['duration(s)']
di = np.diff(dur)
di = np.insert(di, 0, 0)
dur_di = pd.Series(di)
dur_di_freq = dur_di.value_counts()
```

```python
df['time_di'] = time_di
df['dur_di'] = dur_di

# 2. 定义“断裂”的阈值条件
# 例如：时间跳跃大于 1000，或者 dur 发生了重置（小于 0）
break_condition = (df['time_di'] > 1)

# 获取所有满足断裂条件的行索引
break_indices = df[break_condition].index
```

```python
break_indices
```

```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

# 确保 df, break_indices 已经在内存中
# 取出所有断裂点对应的行，按 time_di 降序排列，然后重新提取索引
sorted_break_indices = df.loc[break_indices].sort_values(by='time_di', ascending=False).index

def plot_interactive_break(break_idx_pos=0, plot_window=250, offset=0):
    """
    交互式绘图函数 (双Y轴展示 IED 和 RED)
    """
    if len(sorted_break_indices) == 0:
        print("没有找到符合条件的断裂点。")
        return

    # 获取真实的 DataFrame index
    target_idx = sorted_break_indices[break_idx_pos]
    current_time_di = df.loc[target_idx, 'time_di']
    
    # 计算截取范围并提取数据
    center_idx = target_idx + offset
    start_idx = max(0, center_idx - plot_window)
    end_idx = min(len(df), center_idx + plot_window)
    plot_df = df.iloc[start_idx:end_idx]
    
    # ---------------- 绘图逻辑 (双 Y 轴) ----------------
    fig, ax1 = plt.subplots(figsize=(12, 5))
    
    # 1. 在左侧 Y 轴画 IED (使用蓝色)
    color1 = 'tab:blue'
    ax1.set_xlabel('Sample Sequence (Row Order)')
    ax1.set_ylabel('IED Value (Infrared)', color=color1, fontweight='bold')
    line1 = ax1.plot(range(len(plot_df)), plot_df['ied'], marker='.', markersize=4, linestyle='-', color=color1, label='IED Signal')
    ax1.tick_params(axis='y', labelcolor=color1)
    ax1.grid(True, alpha=0.5)

    # 2. 创建共享 X 轴的右侧 Y 轴 (ax2)
    ax2 = ax1.twinx()  
    
    # 3. 在右侧 Y 轴画 RED (使用红色)
    color2 = 'tab:red'
    ax2.set_ylabel('RED Value (Red Light)', color=color2, fontweight='bold')
    line2 = ax2.plot(range(len(plot_df)), plot_df['red'], marker='.', markersize=4, linestyle='-', color=color2, label='RED Signal', alpha=0.6)
    ax2.tick_params(axis='y', labelcolor=color2)

    # 4. 画断裂点红线 (改用黑色虚线，避免与 RED 信号混淆)
    line_break = []
    if target_idx in plot_df.index:
        relative_pos = plot_df.index.get_loc(target_idx) 
        lb = ax1.axvline(x=relative_pos, color='k', linestyle='--', linewidth=2, label='Break Point')
        line_break = [lb]
    
    # 5. 合并图例 (因为有两个坐标轴，需要把它们收集到一起显示)
    lines = line1 + line2 + line_break
    labels = [l.get_label() for l in lines]
    ax1.legend(lines, labels, loc='upper right') # 统一放在右上角或 'best'
    
    # 标题显示
    plt.title(f'PPG Signals (IED & RED) | Break {break_idx_pos + 1}/{len(sorted_break_indices)} | Row: {target_idx} | time_di = {current_time_di}')
    
    fig.tight_layout() # 调整布局防止标签被遮挡
    plt.show()

# 绑定交互控件
interact(plot_interactive_break, 
         break_idx_pos=widgets.IntSlider(min=0, max=len(sorted_break_indices)-1, step=1, value=0, description='断裂点序号:'),
         plot_window=widgets.IntSlider(min=5, max=1000, step=5, value=250, description='窗口大小:'),
         offset=widgets.IntSlider(min=-800, max=800, step=1, value=0, description='左右平移:'))
```
